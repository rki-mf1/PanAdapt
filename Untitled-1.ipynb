```python
import glob
import re
import pandas as pd

# File pattern and regular expression for date extraction
pattern = 'results/*Delta*/build_site_fubar_table/surface_glycoprotein.fubar.tsv'
date_pattern = re.compile(r'\d{4}-\d{2}')

# Creating a dictionary to store file paths sorted by date
pairs_dict = {}
for path in sorted(glob.glob(pattern, recursive=True)):
    if (date_match := date_pattern.search(path)):
        pairs_dict.setdefault(date_match.group(), []).append(path)

# Keep only dates with exactly two paths
pairs_dict = {date: paths for date, paths in pairs_dict.items() if len(paths) == 2}

# Preparing DataFrames and merging them
dataframes_dict = {}
columns_to_keep = ['Position_in_reference', 'Prob[alpha<beta]']

for date, (path_ja, path_nein) in pairs_dict.items():
    df_ja = pd.read_csv(path_ja, delimiter='\t').drop_duplicates(subset=columns_to_keep[0])[columns_to_keep]
    df_nein = pd.read_csv(path_nein, delimiter='\t').drop_duplicates(subset=columns_to_keep[0])[columns_to_keep]

    merged_df = pd.merge(df_ja, df_nein, on=columns_to_keep[0], how='inner', suffixes=('_ja', '_nein'))

    # Update column names to include date, except for 'Position_in_reference'
    merged_df.columns = [f"{date}_{col}" if col != 'Position_in_reference' else col for col in merged_df.columns]
    dataframes_dict[date] = merged_df

# Extracting positions based on a probability threshold
positions = set()
for merged_df in dataframes_dict.values():
    valid_positions = merged_df[merged_df.filter(like='Prob[alpha<beta]').ge(0.9).any(axis=1)]['Position_in_reference']
    positions.update(valid_positions)

# Filtering DataFrames based on the extracted positions
for date in dataframes_dict:
    dataframes_dict[date] = dataframes_dict[date][dataframes_dict[date]['Position_in_reference'].isin(positions)]

# Merging all DataFrames on 'Position_in_reference'
large_merged_df = pd.DataFrame()
for df in dataframes_dict.values():
    if large_merged_df.empty:
        large_merged_df = df
    else:
        large_merged_df = pd.merge(large_merged_df, df, on='Position_in_reference', how='inner')

# Reset index for the large DataFrame
large_merged_df.reset_index(drop=True, inplace=True)
```

```python
import glob
import re
import pandas as pd
from Bio.Seq import Seq
from collections import Counter

# File pattern and regular expression for date extraction
pattern = 'results/*Delta*/build_site_fubar_table/surface_glycoprotein.fubar.tsv'
date_pattern = re.compile(r'\d{4}-\d{2}')

# Initialize dictionaries to store file paths and codon counters
pairs_dict = {}
seq_df_list = []

for path in sorted(glob.glob(pattern, recursive=True)):
    if (date_match := date_pattern.search(path)):
        date = date_match.group()
        pairs_dict.setdefault(date, []).append(path)

        # Read the file and process Seq_ columns
        df = pd.read_csv(path, delimiter='\t')
        seq_columns = [col for col in df.columns if col.startswith('Seq_')]
        relevant_columns = ['Position_in_reference'] + seq_columns
        seq_df = df[relevant_columns].drop_duplicates(subset='Position_in_reference')

        # Add date information to Seq_ column names
        seq_df.columns = [f"{date}_{col}" if col != 'Position_in_reference' else col for col in seq_df.columns]
        seq_df_list.append(seq_df)

# Keep only dates with exactly two paths
pairs_dict = {date: paths for date, paths in pairs_dict.items() if len(paths) == 2}

# Columns to keep
columns_to_keep = ['Position_in_reference', 'Prob[alpha<beta]']

# Preparing DataFrames and merging them
dataframes_dict = {}
for date, (path_ja, path_nein) in pairs_dict.items():
    df_ja = pd.read_csv(path_ja, delimiter='\t')[columns_to_keep]
    df_nein = pd.read_csv(path_nein, delimiter='\t')[columns_to_keep]

    # Merge and update column names
    merged_df = pd.merge(df_ja, df_nein, on=columns_to_keep[0], how='inner', suffixes=('_ja', '_nein'))
    merged_df.columns = [f"{date}_{col}" if col != 'Position_in_reference' else col for col in merged_df.columns]
    dataframes_dict[date] = merged_df

# Concatenate all Seq_ DataFrame and count codons
all_seq_df = pd.concat(seq_df_list)
seq_codon_counts = {col: all_seq_df.groupby('Position_in_reference')[col].apply(lambda x: Counter(x.dropna())).to_dict() for col in all_seq_df.columns if col != 'Position_in_reference'}

# Extract 'Ref_codons' from one of the files
sample_file = pairs_dict[list(pairs_dict.keys())[0]][0]
ref_codons_df = pd.read_csv(sample_file, delimiter='\t')[['Position_in_reference', 'Ref_codons']].drop_duplicates(subset='Position_in_reference')

# Merging all DataFrames on 'Position_in_reference'
large_merged_df = pd.DataFrame(columns=['Position_in_reference'])
for df in dataframes_dict.values():
    large_merged_df = pd.merge(large_merged_df, df, on='Position_in_reference', how='outer')

# Merge 'Ref_codons' into large_merged_df
large_merged_df = pd.merge(large_merged_df, ref_codons_df, on='Position_in_reference', how='left')

# Translate 'Ref_codons' to amino acids using Biopython
large_merged_df['Ref_amino_acids'] = large_merged_df['Ref_codons'].apply(lambda codon: str(Seq(codon).translate()) if pd.notna(codon) else '')

# Iterate over columns and calculate differences
for col in large_merged_df.columns:
    if col.endswith('_ja'):
        base_col_name = col[:-3]  # Remove '_ja' to get the base name
        nein_col = base_col_name + '_nein'

        if nein_col in large_merged_df.columns:
            diff_col_name = f"{col}_diff"
            large_merged_df[diff_col_name] = large_merged_df[col].sub(large_merged_df[nein_col]).abs()

# Add codon counters to large_merged_df for each yyyy-mm_Prob[alpha<beta]_* column
for col in large_merged_df.columns:
    if 'Prob[alpha<beta]' in col:
        base_col_name = col.split('Prob[alpha<beta]')[0].strip('_')
        for seq_col, counter_dict in seq_codon_counts.items():
            if seq_col.startswith(base_col_name):
                counter_col_name = f"{col}_codon_counter"
                large_merged_df[counter_col_name] = large_merged_df['Position_in_reference'].map(counter_dict)

# Add a column for the maximum diff value
large_merged_df['max_diff'] = large_merged_df[[col for col in large_merged_df.columns if col.endswith('_diff')]].max(axis=1)

# Sort the DataFrame by 'max_diff' in descending order
large_merged_df = large_merged_df.sort_values(by='max_diff', ascending=False)

# Display the DataFrame
print(large_merged_df)
```
