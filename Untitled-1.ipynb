```python
import os
import pandas as pd

# Define the main directory
main_dir = "unfiltered_tsv"

# Create empty dictionaries to store dataframes for each file type
vaccinated_dfs = {}
not_vaccinated_dfs = {}

# Iterate through each subdirectory in the main directory
for subdir in os.listdir(main_dir):
    full_subdir_path = os.path.join(main_dir, subdir)
    
    # Ensure it's a directory
    if os.path.isdir(full_subdir_path):
        year_month = subdir.split('_')[0] + '_' + subdir.split('_')[1]
        
        # Iterate through each file in the subdirectory
        for file in os.listdir(full_subdir_path):
            file_path = os.path.join(full_subdir_path, file)
            df = pd.read_csv(file_path, sep="\t")
            
            # Rename the 'Postmean_W' column to year-month
            df = df[['Position_in_reference', 'Postmean_W']].rename(columns={'Postmean_W': year_month})
            
            # Store the dataframe in the appropriate dictionary
            if "vaccinated" in subdir:
                if file not in vaccinated_dfs:
                    vaccinated_dfs[file] = df
                else:
                    vaccinated_dfs[file] = pd.merge(vaccinated_dfs[file], df, on='Position_in_reference', how='outer')
            else:
                if file not in not_vaccinated_dfs:
                    not_vaccinated_dfs[file] = df
                else:
                    not_vaccinated_dfs[file] = pd.merge(not_vaccinated_dfs[file], df, on='Position_in_reference', how='outer')

# Save the combined data to TSV files
for file, df in vaccinated_dfs.items():
    df.to_csv(os.path.join(main_dir, f"combined_{file}_vaccinated.tsv"), sep="\t", index=False)

for file, df in not_vaccinated_dfs.items():
    df.to_csv(os.path.join(main_dir, f"combined_{file}_not_vaccinated.tsv"), sep="\t", index=False)

import pandas as pd

# Sample DataFrame
df = pd.read_csv('unfiltered_tsv/combined_surface_glycoprotein.tsv_vaccinated.tsv', sep='\t')

# Group by the first column and sum the other columns
grouped_df = df.groupby('Position_in_reference').mean().reset_index()

grouped_df.to_csv('surface_glycoprotein_df.tsv', sep='\t', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the TSV file
df = pd.read_csv('surface_glycoprotein_df.tsv', sep='\t')

# Filter rows based on specified positions
positions_to_consider = [67, 95, 142, 339, 371, 373, 375, 417, 440, 446, 477, 478, 484, 493,  496, 498, 501, 505, 547, 614, 655, 679, 681, 764, 796, 856, 954, 969, 981]
df = df[df['Position_in_reference'].isin(positions_to_consider)].reset_index(drop=True)

# Order of x-axis (Year-Month)
order_x = ['2021-08', '2021-09', '2021-10', '2021-11', '2021-12', '2022-01', '2022-02', '2022-03']

# Rearrange columns based on the desired order
all_columns = ['Position_in_reference'] + [col for month in order_x for col in df.columns if month in col]
df = df[all_columns]

# Separate columns
vaccinated_columns = [col for col in df.columns if 'unvaccinated' not in col and col != 'Position_in_reference']
unvaccinated_columns = [col for col in df.columns if 'unvaccinated' in col and col != 'Position_in_reference']

# Plotting
fig, ax = plt.subplots(figsize=(15, 10))

# Vaccinated
for column in vaccinated_columns:
    ax.scatter([column.split('_')[0]] * len(df), range(len(df)), s=df[column]*20, color='lightblue', alpha=0.5, edgecolors='black')

# Unvaccinated
for column in unvaccinated_columns:
    ax.scatter([column.split('_')[0]] * len(df), range(len(df)), s=df[column]*20, color='lightcoral', alpha=0.5, edgecolors='black')

# Setting y-ticks and y-tick labels
ax.set_yticks(range(len(df)))
ax.set_yticklabels(df['Position_in_reference'])
ax.set_xticks(order_x)
ax.set_xticklabels(order_x, rotation=45)
ax.set_xlabel("Year-Month")
ax.set_ylabel("Position in Reference")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

# Read the TSV file
df = pd.read_csv('surface_glycoprotein_df.tsv', sep='\t')

# Filter rows based on specified positions
positions_to_consider = [67, 95, 142, 339, 371, 373, 375, 417, 440, 446, 477, 478, 484, 493,  496, 498, 501, 505, 547, 614, 655, 679, 681, 764, 796, 856, 954, 969, 981]
df = df[df['Position_in_reference'].isin(positions_to_consider)].reset_index(drop=True)

# Create new columns by subtracting unvaccinated values from vaccinated values for each month
for month in ['2021-08', '2021-09', '2021-10', '2021-11', '2021-12', '2022-01', '2022-02', '2022-03']:
    vaccinated_col = [col for col in df.columns if month in col and 'unvaccinated' not in col][0]
    unvaccinated_col = [col for col in df.columns if month in col and 'unvaccinated' in col][0]
    df[month] = df[vaccinated_col] - df[unvaccinated_col]

# Order of x-axis (Year-Month)
order_x = ['2021-08', '2021-09', '2021-10', '2021-11', '2021-12', '2022-01', '2022-02', '2022-03']

# Plotting
fig, ax = plt.subplots(figsize=(15, 10))

# Plotting the difference with binary coloring
for i, month in enumerate(order_x):
    ax.scatter([month] * len(df), range(len(df)), 
               s=abs(df[month])*20, 
               color=['blue' if val >= 0 else 'red' for val in df[month]], 
               alpha=0.5, edgecolors='black')

# Setting y-ticks and y-tick labels
ax.set_yticks(range(len(df)))
ax.set_yticklabels(df['Position_in_reference'])
ax.set_xticks(order_x)
ax.set_xticklabels(order_x, rotation=45)
ax.set_xlabel("Year-Month")
ax.set_ylabel("Position in Reference")

# Adding a legend
legend_elements = [mpatches.Patch(color='blue', label='More vaccinated'),
                   mpatches.Patch(color='red', label='More unvaccinated')]
ax.legend(handles=legend_elements, loc='upper right')

plt.tight_layout()
# plt.show()
plt.savefig('per_site_vacc_comparison.png')
```
